### 2. Задание (на листочке)
Почему прямые не кажутся перпендикулярными? (см.ролик)

#### Необходимо установить соотношение сторон (осей) координатной плоскости 1:1 с помощью метода plt.axes().set_aspect('equal')
#### По умолчанию стоит режим auto - максимальное заполнение координатной плоскости
#### Ниже графики нарисованы в режиме equal и прямые перпендикулярны

```python
plt.figure(figsize = (5, 5))

x = np.linspace(-15, 15, 21)

y1 =    3*x + 1
y2 = -1/3*x + 1

plt.plot(x, y1)
plt.plot(x, y2)

plt.xlabel('x')
plt.ylabel('y')

plt.axis('tight')
plt.grid(True)
plt.axes().set_aspect('equal')
plt.ylim([-10, 10])

plt.show()
```

### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


#### Окружность

```python
def draw_circle(r):
    
    plt.figure(figsize=(4, 4))

    x = np.linspace(-r, r, 200)
    y1 = np.sqrt(r**2 - x**2)
    y2 = -np.sqrt(r**2 - x**2)

    plt.plot(x, y1, linewidth = 3, linestyle = '-', color = 'darkblue')
    plt.plot(x, y2, linewidth = 3, linestyle = '-', color = 'darkblue')
    
    plt.grid(True)
    plt.xlabel("x")
    plt.ylabel("y")
    
draw_circle(2)
```

#### Эллипс

```python
def draw_ellipse(a, b):
    
    plt.figure(figsize = (6, 4))

    x = np.linspace(-a, a, 1000)
    y1 =  np.sqrt( b**2 * (a**2 - x**2) / a**2 )
    y2 = -np.sqrt( b**2 * (a**2 - x**2) / a**2 ) 

    plt.plot(x, y1, linewidth = 3, linestyle = '-', color = 'darkblue')
    plt.plot(x, y2, linewidth = 3, linestyle = '-', color = 'darkblue')
    
    plt.grid(True)
    plt.xlabel("x")
    plt.ylabel("y")

draw_ellipse(5, 2)
```

#### Гипербола

```python
def draw_hyperbola(a, b):
    
    plt.figure(figsize = (6, 4))

    x = np.linspace(-20, 20, 1000)
    y1 =   (b / a) * np.sqrt( x**2 - a**2 )
    y2 =  -(b / a) * np.sqrt( x**2 - a**2 )

    plt.plot(x, y1, linewidth = 3, linestyle = '-', color = 'darkblue')
    plt.plot(x, y2, linewidth = 3, linestyle = '-', color = 'darkblue')
    
    plt.grid(True)
    plt.xlabel("x")
    plt.ylabel("y")

draw_hyperbola(5, 2)
```

### 4. Задание (на листочке)
1) Пусть задана плоскость: Ax + By + Cz + D = 0

Напишите уравнение плоскости, параллельной данной и проходящей через начало координат.

2)  Пусть задана плоскость: A1x + B1y + C1z + D1 = 0
и прямая: (x-x1)/(x2-x1) = (y-y1)/(y2-y1) = (z-z1)/(z2-z1)

Как узнать, принадлежит прямая плоскости или нет?

#### 4.1 уравнение плоскости, параллельной данной и проходящей через начало координат

если точка (0, 0, 0) лежит на плоскости, то уравнение плоскости имеет вид:
A(x-0) + B(y-0) + C(z - 0) - 0 = 0

Ответ: Ax + By + Cz = 0

#### Как узнать, принадлежит прямая плоскости или нет

Прямая принадлежит плоскости, если координаты любой точки этой прямой удовлетворяют уравнению плоскости.

Соответственно, подставим обе точки прямой (x1, y1, z1) и (x2, y2, z2) и в уравнение плоскости A1x + B1y + C1z + D1 = 0.

Если равенство нулю выполняется, то прямая принадлежит плоскости.

### 5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.

Нарисуйте трехмерный график двух любых поверхностей второго порядка.


#### График двух параллельных плоскостей

```python
fig = figure()
ax = Axes3D(fig)

X = np.arange(-5, 5, 0.5)
Y = np.arange(-10, 10, 0.5)

X, Y = np.meshgrid(X, Y)

Z1 = 7*X + 4*Y
Z2 = 7*X + 4*Y + 100

ax.plot_wireframe(X, Y, Z1, linewidth = 1, color = 'darkred')
ax.plot_wireframe(X, Y, Z2, linewidth = 1, color = 'darkblue')

show()
```

#### График двух любых поверхностей второго порядка

#### Конус

```python
fig = figure()
ax = Axes3D(fig)

X = np.arange(-5, 5, 0.5)
Y = np.arange(-10, 10, 0.5)

X, Y = np.meshgrid(X, Y)

a = 49
b = 9
c = 1

Z1 =  np.sqrt(( a * X**2 + b * Y**2) / c )
Z2 = -np.sqrt(( a * X**2 + b * Y**2) / c )

ax.plot_wireframe(X, Y, Z1, linewidth = 1, color = 'darkblue')
ax.plot_wireframe(X, Y, Z2, linewidth = 1, color = 'darkblue')

show()
```
