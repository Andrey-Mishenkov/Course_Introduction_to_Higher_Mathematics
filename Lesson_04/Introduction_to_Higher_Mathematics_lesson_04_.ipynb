Координаты нижней точки иглы (х,у), игла лежит под углом alfa.  
Пересекает ли игла линию или нет?

Игла пересекает линию, если длина двух отрезков ***h + k >= a***  
* ***h = b * sin(alpha)***  
* ***k = y - a * n***, где ***n = int(y/a)***, целая чать от деления 

Т.о. игла пересекает линию (или касается ее), если выполняется неравенство:

$b * \sin(\alpha) + y - a * int({\frac{y}{a}}) \ge a $
